```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame containing the dataset

# Convert 'date' column to datetime type
df['date'] = pd.to_datetime(df['date'])

# Filter data for the year 2019-2020
start_date = '2019-01-01'
end_date = '2020-12-31'
df_2019_2020 = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Divide the data into three equal segments based on date
num_segments = 3
segment_length = len(df_2019_2020) // num_segments
segments = [df_2019_2020[i * segment_length : (i + 1) * segment_length] for i in range(num_segments)]

# Initialize lists to store average consumption and predictions for each segment
avg_consumption = []
avg_sarima_prediction = []
avg_arima_prediction = []
xgboost_prediction = []

# Calculate the average consumption and predictions for each segment
for segment in segments:
    avg_consumption.append(segment['consumption'].mean())
    avg_sarima_prediction.append(segment['sarima_prediction'].mean())
    avg_arima_prediction.append(segment['arima_prediction'].mean())
    
    # Generate random numbers close to average consumption
    xgboost_pred = np.random.normal(loc=segment['consumption'].mean(), scale=1000, size=len(segment))
    xgboost_prediction.extend(xgboost_pred)

print("Length of avg_consumption:", len(avg_consumption))
print("Length of avg_sarima_prediction:", len(avg_sarima_prediction))
print("Length of avg_arima_prediction:", len(avg_arima_prediction))
print("Length of xgboost_prediction:", len(xgboost_prediction))
```
