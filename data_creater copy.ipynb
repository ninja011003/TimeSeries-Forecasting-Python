```python
import random
import pandas as pd
import numpy as np
import datetime

import matplotlib.pyplot as plt
```

```python
def generate_random_numbers(n, lower_bound, upper_bound):
    random_numbers = [random.randint(lower_bound, upper_bound) for _ in range(n)]
    return random_numbers
def generate_random_floats(n, lower_bound, upper_bound):
    random_numbers = [random.uniform(lower_bound, upper_bound) for _ in range(n)]
    return random_numbers
def random_float_near_n(n, range_width=1.0):
    # Calculate the lower and upper bounds of the range
    lower_bound = n - (range_width / 2)
    upper_bound = n + (range_width / 2)

    # Generate a random float within the specified range
    return random.uniform(lower_bound, upper_bound)
```

```python


def generate_one_year_time_series_with_factors(initial_connection,final_connection, factors, start_year, freq='D'):
    
    if len(factors) != 3:
        raise ValueError("Factors array must contain exactly 3 elements.")

    # Calculate the start and end dates for one year
    start_date = datetime.date(start_year, 1, 1)
    end_date = datetime.date(start_year, 12, 31)

    # Create a date range based on the start_date, end_date, and frequency
    date_range = pd.date_range(start=start_date, end=end_date, freq=freq)
    connections=generate_random_numbers(len(date_range),initial_connection,final_connection)
    
    # Initialize an empty DataFrame
    df = pd.DataFrame({'Date': date_range})

    # Calculate the number of time steps in each segment
    num_steps_per_segment = len(date_range) // 3

    # Initialize the factor index and population
    factor_index = 0
    # initial_factors = generate_random_floats(num_steps_per_segment,factors[0]);

    # Generate population values based on the segmented factors
    connections.sort()
    populations = []
    actual_limits=[]
    supply =[]
    factor=[]

    for i in range(len(date_range)):
        populations.append((int)(connections[i]*4.799))
        actual_limits.append(populations[i]*135)
        factor.append(random_float_near_n(factors[factor_index],0.09))
        supply.append(populations[i]*(135*factor[i]))

        # Move to the next factor segment when needed
        if (i + 1) % num_steps_per_segment == 0 and factor_index < 2:
            factor_index += 1

    # Add the population values to the DataFrame
    df['connection']=connections
    df['population']=populations
    df['actual_limit']=actual_limits
    df['factor']=factor
    df['consumption']=supply
    existing_csv_filename = 'time_series_data.csv'
    existing_data = pd.read_csv(existing_csv_filename)

    # Concatenate the existing data and the additional data
    combined_data = pd.concat([existing_data, df], ignore_index=True)

    # Save the combined data back to the CSV file (mode='w' overwrites, mode='a' appends)
    combined_data.to_csv(existing_csv_filename, index=False, mode='a', header=False)
    return df


```
